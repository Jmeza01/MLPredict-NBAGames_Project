```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier

rr = RidgeClassifier(alpha=1)
split = TimeSeriesSplit(n_splits=3)

sfs = SequentialFeatureSelector(rr, n_features_to_select=30, direction="forward", cv=split)
```

```python
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]
```

```python
selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
sfs.fit(df[selected_columns], df["target"]) # brings both target(actual) column and selected Columns together
```

```python
predictors = list(selected_columns[sfs.get_support()]) # get_support helps to determine what to select for the columns that are best suitable to target from selected columns
```

```python
predictors
```

```python
def backtest(data, model, predictors, start=2, step=1): #start(total of seasons to begin with) step(for one future season at a time)
    all_predictions = [] # list of dataframes for each ..prediction for a single season
    
    seasons = sorted(data["season"].unique())
    
    for i in range(start, len(seasons), step):
        season = seasons[i] # find the index 2 from list of seasons to start with.
        
        train = data[data["season"] < season] # Before index 2 (trains 2016 and 2017 seasons)
        test = data[data["season"] == season] # Same as index 2 (predicts or tests for 2018 season)
        
        model.fit(train[predictors], train["target"] ) # "target column" correlates with "predictors stats" and gets trained with historical data
        
        preds = model.predict(test[predictors]) # generate our predictions only on the test set (Not on the train set)
        # This is where we know how good our model is by using our historical data.
        
        #---by default ---NumPy Array is generated but we are overriding this with pandas.Series()
        
        preds = pd.Series(preds, index=test.index) # Bringing all together including index from test
        
        combined = pd.concat([test["target"], preds], axis=1)
        # create two columns by having axis set at one which are the predictions and target dataframes combined together.
        
        combined.columns = ["actual", "prediction"] # target and predictions looks better
        
        all_predictions.append(combined)
    return pd.concat(all_predictions) # creates rows for all actual and predictions into a long Dataframe 
```

```python
predictions = backtest(df, rr, predictors)
```

```python
from sklearn.metrics import accuracy_score

predictions = predictions[predictions["actual"] != 2]
accuracy_score(predictions["actual"], predictions["prediction"]) # Prediction Percentage we know they Win
```

```python
df_rolling = df[list(selected_columns) + ["won", "team", "season"]] # adding Columns to find the previous 10 games historial data for the teams
```

```python
def find_team_averages(team):
    rolling = team.rolling(10).mean() # means() is the average from current and previous 9 games
    # pandas rolling() to get the previous 10 rows from the team
    return rolling

df_rolling = df_rolling.groupby(["team","season"], group_keys=False).apply(find_team_averages)
```

```python
df_rolling
```

```python
rolling_cols = [f"{col}_10" for col in df_rolling.columns] # To name the columns for previous 10 games
df_rolling.columns = rolling_cols

df = pd.concat([df, df_rolling], axis=1)
```

```python
df
```

```python
df = df.dropna() # drops missing data in rows (Panda feature)
```

```python
df
```

```python
def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_col(df, col_name):
    return df.groupby("team", group_keys=False).apply(lambda x: shift_col(x, col_name))

df["home_next"] = add_col(df,"home")
df["team_opp_next"] = add_col(df,"team_opp")
df["date_next"] = add_col(df,"date")
```
